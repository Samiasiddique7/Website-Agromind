# # # MAE

```python
import matplotlib.pyplot as plt

# Model names
models = ['Simple Linear Regression', 'Multiple Linear Regression', 'Support Vector Regression', 'Decision Tree Regression']

# Corresponding Mean Absolute Error (MAE) values
mae_values = [26.60, 6.51, 72.59, 68.12]

# Specify the desired bar width
bar_width = 0.4  # Adjust this value as needed

# Plotting the bar chart with adjusted bar width
bars = plt.bar(models, mae_values, color=['blue', 'green', 'orange', 'purple'], width=bar_width)

# Removing x-axis ticks and labels
plt.xticks([])

# Adding y-axis label and title
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('Mean Absolute Error Comparison of Regression Models')

# Adding legend
plt.legend(bars, models, loc='upper left', fontsize='small')

# Display the bar chart
plt.show()
```

# # Accuracy

```python
import matplotlib.pyplot as plt

# Model names
models = ['Simple Linear Regression', 'Multiple Linear Regression', 'Support Vector Regression', 'Decision Tree Regression']

# Corresponding Accuracy
accuracy_values = [58.7, 94.2, 96.7, 92.19]

# Specify the desired bar width
bar_width = 0.4  # Adjust this value as needed

# Plotting the bar chart with adjusted bar width
bars = plt.bar(models, accuracy_values, color=['blue', 'green', 'red', 'purple'], width=bar_width)

# Adjusting space between bars
plt.subplots_adjust(wspace=0.4)  # Adjust this value as needed

# Removing x-axis ticks and labels
plt.xticks([])

# Adding y-axis label and title
plt.ylabel('Accuracy %')
plt.title('Accuracy Comparison of Regression Models')

# Adding legend with reduced font size and adjusted position
plt.legend(bars, models, loc='upper left', fontsize='x-small', bbox_to_anchor=(1, 1))

# Display the bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
models = ['Linear', 'Support Vector', 'Decision Tree']
mae_values = [6.01, 72.59, 68.12]
nmse_values = [0.38, 18.29, 14.2]
r2_values = [0.94, 0.97, 0.92]

# Set up the bar chart
bar_width = 0.25
index = np.arange(len(models))

# Plot MAE
plt.bar(index, mae_values, width=bar_width, label='MAE', color='blue')

# Plot N-MSE
plt.bar(index + bar_width, nmse_values, width=bar_width, label='N-MSE', color='orange')

# Plot R2
plt.bar(index + 2 * bar_width, r2_values, width=bar_width, label='R2', color='green')

# Set labels and title
plt.xlabel('Regression Models')
plt.ylabel('Metrics Value')
plt.title('Performance Metrics of Regression Models')
plt.xticks(index + bar_width, models)
plt.legend()

# Show the plot
plt.show()
```
